```python
text = ""
text_f = open('corpus-agriculture-formatted.txt',mode='rt',encoding='utf-8')
for line in text_f:
    text += line.replace(": ", "")
print(len(text))
```

```python
import re
text = re.sub(u"[\u0900-\u097F]", "", text,  flags=re.UNICODE)
text = re.sub(u"[\u0000-\u001F]", "", text,  flags=re.UNICODE)
text = re.sub(u"[\u0021-\u0BFF]", "", text,  flags=re.UNICODE)
text = re.sub(u"[\u0C80-\u10FFFF]", "", text,  flags=re.UNICODE)
```

```python
print(len(text))
```

```python
text_w = open('corpus-agriculture.txt',mode='w+',encoding='utf-8')
text_w.write(text)
```
